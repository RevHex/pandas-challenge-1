## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
#super fluous lineof code uh AVC
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

cd_df['subtotal'] = cd_df.apply((lambda x: x['unit_price']*x['qty']), axis=1)
print (cd_df[['unit_price','qty','subtotal']].head().to_string(index=False), '\n')

# Format Formula to print values as currency.
def cur_formula(value):
    return '${:,.2f}'.format(value)
    
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
cd_df['shipping_weight'] = cd_df.apply(lambda row: row['unit_weight'] * row['qty'], axis=1)
cd_df['shipping_cost'] = cd_df.apply(lambda row: 7.00 * row['shipping_weight']
                                   if row['unit_weight'] * row['qty'] > 50
                                   else 10.00 * row['unit_weight'] * row['qty'],axis=1)
print(cd_df[['unit_price','unit_weight','qty','shipping_weight','shipping_cost']].head().to_string(index=False))
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
sales_tax = 1.0925
cd_df['line_total'] = round(cd_df.apply(lambda row: ((row['subtotal']) + row['shipping_cost'])*sales_tax,axis=1),2)
print(cd_df[['qty','subtotal','shipping_cost','line_total']].head())
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).

cd_df['line_cost'] = cd_df.apply(lambda row: row['unit_cost'] * row['qty'] + row['shipping_cost'], axis=1)
cd_df.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price
cd_df['line_profit'] = cd_df.apply(lambda row: row['line_total'] - row['line_cost'], axis=1)

print(pd.concat([cd_df['order_id'], cd_df[['line_cost', 'line_total', 'line_profit']].map(cur_formula)], axis=1).head().to_string(index=False))
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
top5_clients = cd_df.groupby(['client_id']).size().nlargest(5).index.tolist()
for client_id in top5_clients:
    cust_order_total = cd_df.loc[cd_df['client_id'] == client_id]['line_total'].sum()
    print (f'Client {client_id} Total ${cust_order_total:.2f}')
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
summary_df = pd.DataFrame(columns=['client_id', 'total_qty', 'total_shipping', 'total_revenue', 'total_cost', 'total_profit'])
    #Calculate totals for the Top 5 Client IDs
for client_id in top5_clients:
    total_qty = cd_df.loc[cd_df['client_id'] == client_id]['qty'].sum()
    total_shipping = cd_df.loc[cd_df['client_id'] == client_id]['shipping_cost'].sum()
    total_revenue = cd_df.loc[cd_df['client_id'] == client_id]['line_total'].sum()
    total_cost = cd_df.loc[cd_df['client_id'] == client_id]['line_cost'].sum()
    total_profit = cd_df.loc[cd_df['client_id'] == client_id]['line_profit'].sum()
    #New DF for current client list
    next_df = pd.DataFrame({'client_id'      : [client_id],
                            'total_qty'      : [total_qty],
                            'total_shipping' : [total_shipping],
                            'total_revenue'  : [total_revenue],
                            'total_cost'     : [total_cost],
                            'total_profit'   : [total_profit]})

    #Concatenate new DF with Summary DF
    summary_df = pd.concat([summary_df, next_df], ignore_index=True, axis=0)

#Sort the Summary DF by total_profit as indicated above
summary_df.sort_values(by='total_profit',ascending=False)
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
summary_df = summary_df.rename(columns = {'client_id'      : 'Client ID',
                                          'total_qty'      : 'Quantity',
                                          'total_shipping' : 'Shipping',
                                          'total_revenue'  : 'Total Revenue',
                                          'total_cost'     : 'Total Cost',
                                          'total_profit'   : 'Total Profit'
})
# Format results so that currency is in millions of dollars
presentation_df = summary_df.assign(**{
    'Shipping'      : lambda x: ('$' + (x['Shipping']      /1_000_000).round(2).astype(str) + 'M'),
    'Total Revenue' : lambda x: ('$' + (x['Total Revenue'] /1_000_000).round(2).astype(str) + 'M'),
    'Total Cost'    : lambda x: ('$' + (x['Total Cost']    /1_000_000).round(2).astype(str) + 'M'),
    'Total Profit'  : lambda x: ('$' + (x['Total Profit']  /1_000_000).round(2).astype(str) + 'M')
})
presentation_df
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
presentation_df.sort_values(by='Total Profit', ascending=False)
```
